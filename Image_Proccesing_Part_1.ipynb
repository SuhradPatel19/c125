```python
from matplotlib import pyplot
from matplotlib.image import imread

trainingInfectedImage='/content/PRO-M3-Pneumothorax-Image-Dataset/training_dataset/infected/image_1.png'

image=imread(trainingInfectedImage)
print(image)

pyplot.title("Infected Image")
pyplot.imshow(image)
pyplot.show()
```

```python
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# argumenting images by using ImageDataGenerator
trainingDataGenerator=ImageDataGenerator(
    rotation_range=180,width_shift_range=0.3,fill_mode="nearest"
)

trainingImageDirt="/content/PRO-M3-Pneumothorax-Image-Dataset/training_dataset"

print(trainingImageDirt)

print(trainingDataGenerator)

my_training_argumented_images=trainingDataGenerator.flow_from_directory(trainingImageDirt,target_size=(180,180))
print(my_training_argumented_images)
```

```python
for i in range(10):
  pyplot.subplot(2,5,i+1)
  myimages=my_training_argumented_images.next()
  # print(myimages)

  #print(len(myimages))

  finalImage=myimages[0][0].astype("uint8")
  print(finalImage)



  pyplot.imshow(finalImage)
```
